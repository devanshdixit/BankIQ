# EDA on Enhanced Customer Data

```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Load the enhanced data
enhanced_customers = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/processed/enhanced_customers.csv")
print("Enhanced data loaded successfully!")
```

```python

# Basic information and summary
print("Enhanced Customers Data Info:")
print(enhanced_customers.info())
print(enhanced_customers.describe())

# Display first few rows
print("Sample Data:")
print(enhanced_customers.head())
```

```python

# Age Group Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=enhanced_customers, x="AgeGroup", palette="viridis")
plt.title("Age Group Distribution")
plt.show()
```

```python

# Income Bracket Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=enhanced_customers, x="IncomeBracket", palette="Set2")
plt.title("Income Bracket Distribution")
plt.show()
```

```python

# Risk Score Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=enhanced_customers, x="RiskScore", palette="coolwarm")
plt.title("Credit Risk Score Distribution")
plt.show()
```

```python

# Average Transaction Amount
plt.figure(figsize=(8, 5))
sns.histplot(enhanced_customers["AvgTransactionAmount"], bins=30, kde=True, color='orange')
plt.title("Average Transaction Amount Distribution")
plt.xlabel("Amount")
plt.show()
```

```python

# Product Count Distribution
plt.figure(figsize=(8, 5))
sns.histplot(enhanced_customers["ProductCount"], bins=10, kde=True, color='blue')
plt.title("Product Count per Customer")
plt.xlabel("Number of Products")
plt.show()
```

```python

# EMI to Income Ratio Distribution
plt.figure(figsize=(8, 5))
sns.histplot(enhanced_customers["AvgEMItoIncomeRatio"], bins=20, kde=True, color='purple')
plt.title("EMI to Income Ratio Distribution")
plt.xlabel("EMI to Income Ratio")
plt.show()
```

```python

# High-Risk Loan Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=enhanced_customers, x="HighRiskLoan", palette="Set1")
plt.title("High-Risk Loan Distribution")
plt.xlabel("High-Risk Loan (1=Yes, 0=No)")
plt.show()
```

```python
# Filter only numeric columns for correlation
numeric_cols = enhanced_customers.select_dtypes(include=[np.number]).columns

# Correlation Heatmap of Numeric Features
plt.figure(figsize=(12, 8))
corr_matrix = enhanced_customers[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Enhanced Features")
plt.show()
```
