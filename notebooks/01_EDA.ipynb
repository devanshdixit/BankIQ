1. Setting Up the Environment:

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Load cleaned data
customers = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_customers.csv")
products = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_products.csv")
transactions = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_transactions.csv")
loans = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_loans.csv")
campaign_responses = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_campaign_responses.csv")
support_interactions = pd.read_csv("/Users/devanshudixit/Desktop/projects/BankIQ/data/cleaned/cleaned_support_interactions.csv")

print("Data loaded successfully!")
```

2. Basic Data Overview:

```python
# Display basic information
print("\nCustomers Data:")
print(customers.info())
print(customers.describe())

print("\nProducts Data:")
print(products.info())
print(products.describe())

print("\nTransactions Data:")
print(transactions.info())
print(transactions.describe())

print("\nLoans Data:")
print(loans.info())
print(loans.describe())

print("\nCampaign Responses Data:")
print(campaign_responses.info())

print("\nSupport Interactions Data:")
print(support_interactions.info())
```

3. Distribution Analysis:

```python
# Customer Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(customers["Age"], bins=20, kde=True, color='blue')
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
# Income Distribution:
plt.figure(figsize=(8, 5))
sns.histplot(customers["Income"], bins=30, kde=True, color='green')
plt.title("Income Distribution")
plt.xlabel("Annual Income")
plt.show()
```

4. Categorical Analysis:

```python
# Gender Distribution:
plt.figure(figsize=(6, 4))
sns.countplot(data=customers, x="Gender", palette="viridis")
plt.title("Gender Distribution")
plt.show()
```

```python
# Product Type Analysis:
plt.figure(figsize=(10, 5))
sns.countplot(data=products, x="ProductType", palette="Set2")
plt.title("Product Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

5. Financial Insights:

```python
# Credit Score Distribution:
plt.figure(figsize=(8, 5))
sns.histplot(customers["CreditScore"], bins=20, kde=True, color='purple')
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.show()
```

```python
# Balance Distribution by Product:
plt.figure(figsize=(10, 6))
sns.boxplot(data=products, x="ProductType", y="Balance", palette="coolwarm")
plt.title("Balance Distribution by Product Type")
plt.xticks(rotation=45)
plt.show()
```

6. Transaction Analysis:

```python
# Transaction Amount Distribution:
plt.figure(figsize=(8, 5))
sns.histplot(transactions["Amount"], bins=30, kde=True, color='orange')
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.show()
```

```python
# Transaction Types:
plt.figure(figsize=(8, 5))
sns.countplot(data=transactions, x="TransactionType", palette="pastel")
plt.title("Transaction Types")
plt.show()
```

7. Loan Insights:

```python
# Loan Amount by Type:
plt.figure(figsize=(10, 6))
sns.boxplot(data=loans, x="LoanType", y="Amount", palette="Blues")
plt.title("Loan Amount Distribution by Loan Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Interest Rate Distribution:
plt.figure(figsize=(8, 5))
sns.histplot(loans["InterestRate"], bins=20, kde=True, color='red')
plt.title("Interest Rate Distribution")
plt.xlabel("Interest Rate (%)")
plt.show()
```

8. Correlation Analysis:

```python
# Correlation between Income, Credit Score, and Age:
plt.figure(figsize=(8, 6))
corr_matrix = customers[["Age", "Income", "CreditScore"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation between Age, Income, and Credit Score")
plt.show()
```

9. Customer Support Insights:

```python
# Support Issue Types:
plt.figure(figsize=(8, 5))
sns.countplot(data=support_interactions, x="IssueType", palette="muted")
plt.title("Support Issue Types")
plt.show()
```

```python
# NPS Score Distribution:
plt.figure(figsize=(8, 5))
sns.histplot(support_interactions["NPSScore"], bins=10, kde=True, color='teal')
plt.title("NPS Score Distribution")
plt.xlabel("NPS Score")
plt.show()
```

10. Campaign Response Analysis:

```python
# Response Types:
plt.figure(figsize=(6, 4))
sns.countplot(data=campaign_responses, x="Response", palette="Set1")
plt.title("Campaign Response Distribution")
plt.show()
```
